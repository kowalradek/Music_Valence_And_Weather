```python
import pandas as pd
from scipy import stats
import statsmodels.formula.api as smf

# Load the cleaned dataset
df = pd.read_csv('data/enriched_music_weather.csv')

# T-Test
rainy_days = df[df['is_raining'] == 'Yes']['valence']
dry_days = df[df['is_raining'] == 'No']['valence']

t_stat, p_val = stats.ttest_ind(rainy_days, dry_days, equal_var=False)
print(f"Rain T-Test -> T-Stat: {t_stat:.4f}, P-Value: {p_val:.4f}")
print(f"Avg Valence (Rain): {rainy_days.mean():.4f} | Avg Valence (Dry): {dry_days.mean():.4f}\n")

# Pearson corrs
metrics = ['temperature_2m_mean', 'sunshine_duration', 'precipitation_sum']

print("Global Correlations with Valence:")
for metric in metrics:
    # Drop NaNs to avoid correlation errors
    clean_subset = df[['valence', metric]].dropna()
    corr, p = stats.pearsonr(clean_subset['valence'], clean_subset[metric])
    print(f"- {metric}: Corr = {corr:.4f} (p={p:.4f})")
print("\n")

# MODERATION OLS Regression
model = smf.ols('valence ~ temperature_2m_mean * C(region)', data=df).fit()

print("Moderation Analysis Summary (Temperature x Region):")

print(model.summary().tables[1])
```
